## 1. Introduction
In the last few months of 2019, the COVID-19 pandemic caused the entire world to adapt. Due to the virus, meetings and lectures were hosted online, and people were not allowed to go to offices. The use of public transport and motorways dropped accordingly. 

This report will explore the impact of the COVID-19 pandemic on the usage of these modes of transport, the recovery in the years after, and a forecast of transport use in the future years.
Objectives
The main objective of this study is to analyze how the performance of different transport modes in the Netherlands was affected by the COVID-19 pandemic and how they have recovered since. The study aims to provide insights into:
- The relative impact of COVID-19 on different modes of passenger transport (car, local public transport (bus, tram, metro), trains, bicycle, and walking).
- Regional variations in recovery between urban and rural areas.
- Regional variations in recovery per province
- Forecasting future transport performance and identifying when levels are expected to return to pre-COVID (2019) levels.

## 2. Research Questions
Main question:\
How were different transport mode performances affected, did they recover after COVID-19 and how was the recovery process?

Sub-questions:
- How did the transport performance of different modes of transport change during and after the corona period (2020–2022) compared to 2019?
- Which transport modes show the largest recovery compared to pre-COVID levels?
- Are there significant differences in the degree of recovery between urban and rural regions in the Netherlands?
- Are there significant differences in the degree of recovery between provinces in the Netherlands?
- Based on historical trends, how is transport performance likely to develop over the next five years (2025–2030), and if or when will each mode of transport return to pre-COVID levels?

## 3. Answering the research questions
#### 3.1 How did the transport performance of different modes of transport change during and after the corona period (2020-2022) compared to 2019?

As seen in figure 1 the transport performance of the various transport modes experienced a significant dip between 2019 and 2020. This was due to the emergency policies and regulations introduced by the Dutch government to suppress the spread of COVID-19 in the Netherlands. Because of the policies most people had to work from home and it was not allowed to visit family and friends. After the corona period, transport performance started to recover again as more and more people returned back to normal everyday life. The total transport performance however never fully recovered to the values before COVID-19 indicating that some people never returned to their original schedule.

#### 3.2 Which transport modes show the largest recovery compared to pre-COVID levels?

In contrast to figure 1, figure 2 shows the relative change in transport performance per travel mode. Figure 2 shows clearly that every transport mode experienced a decrease in performance except for walking. This is because walking became a popular activity for people during the corona period to still go outside and have some form of exercise. It was also an activity for people to break the long days of only staying inside their own homes.

#### 3.3 Are there significant differences in the degree of recovery between urban and rural regions in the Netherlands?

Both graphs show that travel performance dropped sharply in 2020 due to COVID-19 and then slowly recovered between 2021 and 2023. Figure 3 shows similar patterns for all urbanisation levels, but highly urbanised areas recovered a bit faster.

For Figure 4, the categories were grouped as follows:
“Extremely urbanised,” “Strongly urbanised,” and “Moderately urbanised” were combined into Urban, while “Hardly urbanised” and “Not urbanised” were combined into Rural.

Figure 4 confirms this: urban areas show slightly higher recovery levels than rural areas after 2020. The difference is small, so the recovery trend is mostly the same across the country.

To awnser the question:

Urban regions recovered a bit quicker than rural ones, but the overall recovery pattern in the Netherlands is very similar for all regions.
