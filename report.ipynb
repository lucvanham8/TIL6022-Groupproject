## 1. Introduction
In the last few months of 2019, the COVID-19 pandemic caused the entire world to adapt. Due to the virus, meetings and lectures were hosted online, and people were not allowed to go to offices. The use of public transport and motorways dropped accordingly. 

This report will explore the impact of the COVID-19 pandemic on the usage of these modes of transport, the recovery in the years after, and a forecast of transport use in the future years.
Objectives
The main objective of this study is to analyze how the performance of different transport modes in the Netherlands was affected by the COVID-19 pandemic and how they have recovered since. The study aims to provide insights into:
- The relative impact of COVID-19 on different modes of passenger transport (car, local public transport (bus, tram, metro), trains, bicycle, and walking).
- Regional variations in recovery between urban and rural areas.
- Regional variations in recovery per province
- Forecasting future transport performance and identifying when levels are expected to return to pre-COVID (2019) levels.

#### 1.2 Data Sources
CBS Open data StatLine – Totale vervoersprestatie in Nederland; vervoerwijzen, regio’s
	https://opendata.cbs.nl/statline/#/CBS/nl/dataset/84687NED/table?ts=1759138063

The CBS dataset 84687NED – Transport Performance in the Netherlands provides annual figures on the total number of kilometers traveled within the Netherlands, broken down by transport mode (car, public transport, bicycle, freight/other) and by region (provinces and degree of urbanization). The data, collected through the ODiN survey, covers residents aged six and older. In our study, we will use this dataset to compare transport performance across modes before, during, and after COVID-19 with 2019 as a baseline, to assess regional differences between urban and rural areas and to examine recovery trends over time in order to forecast developments up to 2030.

CBS Mobiliteit; per persoon, persoonskenmerken, motieven en regio’s (woon-werk verkeer)
https://opendata.cbs.nl/statline/#/CBS/nl/dataset/84713NED/table

The CBS dataset 84713NED – Mobility by person, personal characteristics, motives and regions provides detailed data on individual mobility behavior in the Netherlands, including the number of trips, distances, transport modes, motives (such as commuting, shopping, leisure), and breakdowns by personal and regional characteristics. In our study we will focus especially on home-work (commuting) travel data: how many trips or kilometers are made for commuting, what transport modes are used, and how this varies across regions and over time.

## 3. Answering the research questions
#### 3.1 How did the transport performance of different modes of transport change during and after the corona period (2020-2022) compared to 2019?
As seen in figure 1 the transport performance of the various transport modes experienced a significant dip between 2019 and 2020. This was due to the emergency policies and regulations introduced by the Dutch government to suppress the spread of COVID-19 in the Netherlands. Because of the policies most people had to work from home and it was not allowed to visit family and friends. After the corona period, transport performance started to recover again as more and more people returned back to normal everyday life. The total transport performance however never fully recovered to the values before COVID-19 indicating that some people never returned to their original schedule.

## 4. Conclusion
After looking into five sub-questions the information to answer the main research question has been acquired. First of all, after plotting the data for the transport performance of each mode it is clear that almost every transport mode experienced a massive decrease in total kilometers travelled due to the pandemic. The only mode that didn’t experience a decrease but an increase in kilometers travelled was walking. This can be explained with the fact that most people had to work from home during the pandemic causing a massive decrease in work related travel. The pandemic also made it harder to practice sports as distance between individuals had to be maintained and inside sport areas were closed, which caused people to go on daily walks for exercise and leisure. 

Secondly figure 2 shows that public transport in the form of trains, buses, trams and metros have taken the longest time to recover. Public transport also took the biggest hit in transport performance because PT companies faced big challenges on how to transport passengers while complying to all COVID-19 regulations.

Lastly the recovery process started off slow as it took people a while to get into the normal course of events after the pandemic. Per province there are significant differences in the recovery process indicating that geographic differences have an impact on how the transport performances recovered after COVID-19. When forecasting the transport performance in the future the the models show that the total transport performance will fully recover back to levels before COVID-19 and then keep on growing.
